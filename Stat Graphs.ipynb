Here I will be making the graphs based on MLB Stats

Sources:
http://proxy.espn.com/mlb/stats/parkfactor?sort=HRFactor


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
HRParkFactor=[1.377,1.292,1.265,1.175,1.168,1.158,1.170,1.130,1.128,1.086,1.081,1.062,1.048,1.030,1.018,1.000,0.934,0.906,0.904,0.888,0.878,0.871,0.871,0.870,0.865,0.854,0.831,0.828,0.741,0.691]
Average=np.sum(HRParkFactor)/30
print(Average)
```

```python
Teamname=['Blue Jays','Nationals','Rockies','Oriels','White Sox','Astros','Phillies','Reds','Dodgers','Tigers','Angels','Rangers','Indians','Brewers','Braves','Mets','Mariners','Pirates','Rays','DiamondBacks','Red Sox','Cubs','Padres','Twins','Yankees','Athletics','Cardinals','Marlins','Royals','Giants']
```

```python
y_pos=np.arange(len(Teamname))
```

```python
plt.figure(num=None, figsize=(10,10), dpi=300, facecolor='w', edgecolor='k')
plt.barh(y_pos,HRParkFactor,align='center',alpha=1,)
plt.yticks(y_pos,Teamname)
plt.axvline(x=Average, color='purple',label= 'League Average')
plt.legend()
plt.xlabel('Home Run Park Factor')
plt.title('Home Run Park Factor by Team')
plt.savefig('HRFactor')
```
